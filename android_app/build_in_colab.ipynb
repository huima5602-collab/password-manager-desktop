# 密码生成与管理器 - Android APK 编译
### 使用 Google Colab 免费编译 APK

**步骤**：
1. 先把 `android_app` 文件夹打包成 `android_app.zip`
2. 上传到 Google Drive，放到根目录
3. 依次运行下面每个代码块

```
# 1. 安装 Buildozer
!sudo apt-get update -qq
!sudo apt-get install -y -qq python3-pip openjdk-17-jdk-headless git wget unzip zip autoconf automake libtool libffi-dev libssl-dev
!pip3 install --upgrade buildozer cython virtualenv
print('环境就绪！')
```

```
# 2. 挂载 Google Drive
from google.colab import drive
drive.mount('/content/drive')
```

```
# 3. 解压项目文件
import zipfile, os

zip_path = '/content/drive/MyDrive/android_app.zip'

if not os.path.exists(zip_path):
    print('请先将 android_app.zip 上传到 Google Drive 根目录！')
else:
    with zipfile.ZipFile(zip_path, 'r') as z:
        z.extractall('/content/pwdapp')
    print('解压完成！文件列表：')
    for root, dirs, files in os.walk('/content/pwdapp'):
        for f in files:
            print(f'  {os.path.join(root, f)}')
```

```
# 4. 编译 APK
import os, time

os.chdir('/content/pwdapp/android_app')
print('开始编译 APK...（首次约 15-30 分钟）')
start = time.time()
!buildozer android debug
elapsed = time.time() - start
print(f'\n耗时: {elapsed/60:.1f} 分钟')
```

```
# 5. 下载 APK
import glob, shutil

apk_files = glob.glob('/content/pwdapp/android_app/bin/*.apk')
if apk_files:
    apk = apk_files[0]
    dest = '/content/drive/MyDrive/密码生成与管理器.apk'
    shutil.copy(apk, dest)
    print(f'APK 已保存到 Google Drive: {dest}')
    print(f'文件大小: {os.path.getsize(apk)/1024/1024:.1f} MB')
else:
    print('未找到 APK，编译可能失败了，请检查上方输出')
```
